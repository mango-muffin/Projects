```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

## Sample Plot 

A scatter plot with points in each of the 4 cartesian quadrants. Each quadrant is assigned a different colour.
<br>
2 lines are also added to designate the quadrant boundaries.
<br> 
This is to test **Markdown** and inline *matplotlib*


```python
# inline means that the plot output will be displayed after the cell and stored in the notebook.
%matplotlib inline

# Creating a default generator for each of the points
rng = np.random.default_rng()

# Each quadrant will have 20 random points. 
# The range of values generated by default_rng is low inclusive, high exclusive, i.e. [low, high).
# The range for the x & y values of the sample plot = [-100, 100).
red = pd.DataFrame(rng.integers(100, size = (20, 2)))
blue = pd.DataFrame({0: rng.integers(-100, 0, size = 20), 
                    1: rng.integers(100, size = 20)})
green = pd.DataFrame(rng.integers(-100, 0, size = (20, 2)))
yellow = pd.DataFrame({0: rng.integers(100, size = 20), 
                    1: rng.integers(-100, 0, size = 20)})

# Set up for legend and colours
group = (red, blue, green, yellow)
colours = ("red", "blue", "green", "gold")
legend = ("1st Quadrant", "2nd Quadrant", "3rd Quadrant", "4th Quadrant")

for data, colour, name in zip(group, colours, legend):
    plt.scatter(data[0], data[1], c = colour, label = name) 

plt.axhline(y = 0, lw = 1) #lw refers to the width of the line
plt.axvline(x = 0, lw = 1)

# Placing the legend outside of the plot so it does not overlap with the points
plt.legend(loc = [1.1, 0.35]) 
plt.show()
```
